# Evaluate the systematic error in the circular velocity in Eilers et al. (2019)

## https://ui.adsabs.harvard.edu/abs/2019ApJ...871..120E/abstract 

## Here I check the systematic error in the circular velocity in Eilers et al. (2019). The goal is to get some insights into my hierarchical Bayesian model. 

```python
import numpy

import matplotlib
from matplotlib import pyplot
from matplotlib.ticker import MaxNLocator
pyplot.rcParams['ps.useafm'] = True
pyplot.rcParams['pdf.use14corefonts'] = True
pyplot.rcParams['text.usetex'] = True
```

# Jeans equation

$$
v_c^2(R) = \langle v_\phi^2 \rangle - \langle v_R^2 \rangle \left[ 1 + \frac{\partial \ln \nu}{\partial \ln R} + \frac{\partial \ln \langle v_R^2 \rangle}{\partial \ln R} \right]
$$

Density
$$
\nu = \nu_0 \exp \left[ - \frac{R}{R'_\mathrm{exp}} \right] \;R'_\mathrm{exp}=21\; \mathrm{kpc}
$$

Velocity dispersion
$$
\langle v_R^2 \rangle = \sigma_{R0}^2 \exp \left[ - \frac{2R}{R_\mathrm{exp}} \right], 
\;\sigma_{R0}=56 \;\mathrm{km\;s^{-1}},  
\;R_\mathrm{exp}=(3 \pm 1)\; \mathrm{kpc}
$$

## Their estimate of $v_c(R)$ is available from their Table 1.

```python
# Load the table
filename_RC    =  'data/R_vc_sigN_sigP__Eilers2019Table1.txt'
data_RC    = numpy.loadtxt(filename_RC)

data_R        = data_RC[:,0]
data_vcirc    = data_RC[:,1]
data_error_RC =(data_RC[:,2]+data_RC[:,3])/2. 
```

## As mentioned in Eilers et al. (2019), the dominant source of uncertaintiy in $v_\mathrm{circ}$ arises from the uncertainty in $R_\mathrm{exp}$. I will check how $v_\mathrm{circ}$ is affected by the change in $R_\mathrm{exp}$.

```python
R_exp_0 = 3.
R_nu_exp_0 = 21.
sig_R_0 = 56.

def vcirc_Rexp(R_exp_model):
    
    vR2 = sig_R_0**2 * numpy.exp(-2.*data_R/R_nu_exp_0) 
    vphi2 = data_vcirc**2 + vR2 * (1 - data_R/R_exp_0 - 2.*data_R/R_nu_exp_0)
    
    vcirc2_model = vphi2 - vR2*(1 - data_R/R_exp_model - 2.*data_R/R_nu_exp_0)
    
    return numpy.sqrt(numpy.fabs(vcirc2_model))
    
```

```python
N_sample = 20

numpy.random.seed(123)
#list_Rexp = R_exp_0 + numpy.random.randn(N_sample)*1
list_Rexp = R_exp_0 + numpy.linspace(-1,1,21)

for i in range(N_sample):
    vcirc_model = vcirc_Rexp(list_Rexp[i])
    if (list_Rexp[i]>=R_exp_0):
        pyplot.plot(data_R, vcirc_model-data_vcirc, '-')
    else:
        pyplot.plot(data_R, vcirc_model-data_vcirc, '--')


pyplot.annotate('dashed: $R_\mathrm{exp}<3\;\mathrm{kpc}$', xy=(18, 4.2), fontsize=13)
pyplot.xticks(fontsize=20)
pyplot.yticks(fontsize=20)
pyplot.xlabel(r'$R / \mathrm{kpc}$', fontsize=20)
pyplot.title(r'$[v_\mathrm{circ}(R|R_\mathrm{exp}) - v_\mathrm{circ}(R|R_\mathrm{exp}=3 \;\mathrm{kpc})] / (\mathrm{km\;s^{-1}})$', fontsize=20)
pyplot.show()
```

```python
N_sample = 20


for i in range(N_sample):
    vcirc_model = vcirc_Rexp(list_Rexp[i]) 
    if (list_Rexp[i]>=R_exp_0):
        pyplot.plot(data_R, vcirc_model/data_vcirc, '-')
    else :
        pyplot.plot(data_R, vcirc_model/data_vcirc, '--')


pyplot.annotate('dashed: $R_\mathrm{exp}<3\;\mathrm{kpc}$', xy=(10, 1.021), fontsize=13)

pyplot.xticks(fontsize=20)
pyplot.yticks(fontsize=20)
pyplot.xlabel(r'$R / \mathrm{kpc}$', fontsize=20)
pyplot.title(r'$v_\mathrm{circ}(R|R_\mathrm{exp}) / v_\mathrm{circ}(R|R_\mathrm{exp}=3 \;\mathrm{kpc})$', fontsize=20)
pyplot.show()
```

## As seen from these plots, the systematic error in $R_\mathrm{exp}$ affects $v_\mathrm{circ}$ by a few km/s at each radius.  
## If $R_\mathrm{exp}<3 \;\mathrm{kpc}$, $v_\mathrm{circ}$ becomes larger than the value presented in Eilers et al. (2019).

```python
# randomly choose integer
N_sample=1000

numpy.random.seed(123)
list_Rexp = R_exp_0 + numpy.random.randn(N_sample)*1

matrix_vcirc  = numpy.empty((len(data_R),N_sample),numpy.float64)
for i in range(N_sample):
    
    vcirc_model = vcirc_Rexp(list_Rexp[i]) 

    #print(list_vcirc.shape)
    matrix_vcirc  [:,i]=vcirc_model



vcirc_p16=numpy.array([])
vcirc_p50=numpy.array([])
vcirc_p84=numpy.array([])

for i in range(len(data_R)):
    vcirc_p16 = numpy.append(vcirc_p16, numpy.percentile(matrix_vcirc[i,:],16) )
    vcirc_p50 = numpy.append(vcirc_p50, numpy.percentile(matrix_vcirc[i,:],50) )
    vcirc_p84 = numpy.append(vcirc_p84, numpy.percentile(matrix_vcirc[i,:],84) )
    
    
pyplot.plot(data_R, vcirc_p16, 'r--')
pyplot.plot(data_R, vcirc_p50, 'k-')
pyplot.plot(data_R, vcirc_p84, 'b--')

pyplot.xticks(fontsize=20)
pyplot.yticks(fontsize=20)
pyplot.xlabel(r'$R / \mathrm{kpc}$', fontsize=20)
pyplot.ylabel(r'$v_\mathrm{circ}(R|R_\mathrm{exp}) / (\mathrm{km\;s^{-1}})$', fontsize=20)

pyplot.show()


    
pyplot.plot(data_R, (vcirc_p84 - vcirc_p16)/2, 'k--')
pyplot.xlabel(r'$R / \mathrm{kpc}$', fontsize=20)

pyplot.xticks(fontsize=20)
pyplot.yticks(fontsize=20)
pyplot.xlabel(r'$R / \mathrm{kpc}$', fontsize=20)
pyplot.ylabel(r'Mean sys. error in $v_\mathrm{circ}(R|R_\mathrm{exp})$', fontsize=20)
pyplot.title(r'$(p_{84}-p_{16})/(2 p_{50})$', fontsize=20)



pyplot.show()


pyplot.plot(data_R, (vcirc_p84 - vcirc_p50), 'b--')
pyplot.plot(data_R, (vcirc_p16 - vcirc_p50), 'r--')
pyplot.xlabel(r'$R / \mathrm{kpc}$', fontsize=20)

pyplot.xticks(fontsize=20)
pyplot.yticks(fontsize=20)
pyplot.xlabel(r'$R / \mathrm{kpc}$', fontsize=20)
pyplot.ylabel(r'Systematic error in $v_\mathrm{circ}(R|R_\mathrm{exp})$', fontsize=20)

pyplot.show()




    
pyplot.plot(data_R, (vcirc_p84 - vcirc_p50)/(vcirc_p50), 'b--')
pyplot.plot(data_R, (vcirc_p16 - vcirc_p50)/(vcirc_p50), 'r--')

pyplot.xticks(fontsize=20)
pyplot.yticks(fontsize=20)
pyplot.xlabel(r'$R / \mathrm{kpc}$', fontsize=20)
pyplot.ylabel(r'Fractional error in $v_\mathrm{circ}(R|R_\mathrm{exp})$', fontsize=20)

pyplot.show()


```
